```python
# 소수 구하기

def get_primes(k) :
    dic = {0 : False, 1:False, 2:True}
    primes = [2]
    for n in range(3, k+1) : 
        for p in primes :
            if n%p == 0 :
                dic[n] = False
                break
            if p == primes[-1] : 
                dic[n] = True
                primes.append(n)
                break
    return dic
```

```python
# 소수인지 확인하기

def is_prime(n) : 
    if n < 2 : return False
    
    for idx in range(2, int(n**0.5)+1) : 
        if n%idx == 0 : return False
    return True

for idx in range(2, 20) :
    print(idx, is_prime(idx))
```

```python
# 리스트의 조합(combination) 구하기
from itertools import chain, combinations

def get_combination(lst):
    return list(chain.from_iterable(combinations(lst, r) for r in range(len(lst)+1)))

get_combination([1,2,3])
```

```python
# 리스트의 순열(permutation) 구하기
from itertools import permutations

def get_permutation(lst) :
    return list(permutations(lst))

get_permutation([1,2,3])
```

```python
# 하나의 리스트에서 가능한 모든 조합의 순열 구하기
from itertools import permutations

def get_all_permutation(lst) :
    result = []
    for r in range(len(lst)+1) :
        result.extend(permutations(lst, r))
    
    return list(set(result))

sorted(get_all_permutation([1,1,2,3]), key=lambda x : len(x))
```

```python
# 두 개 이상의 리스트의 조합 구하기
from itertools import product

def get_product(lst) :
    return list(product(*lst))

get_product([['1','2'], ['3', '4', '5']])
```

```python
# 리스트에서 원소가 중복되지 않도록 원하는 길이의 배열 만들기
from itertools import permutations

def get_all_permutation2(lst, length) :
    result = permutations(lst, length)
    return list(set(result))

sorted(get_all_permutation2([1,2,3,4,5,6,7,8,9], 3))
```

```python
# comparator
from functools import cmp_to_key

def comparator_ascending(x,y) :
    if x<y : return -1
    elif x>y : return 1
    else : return 0
    
def comparator_descending(x,y) :
    if x<y : return 1
    elif x>y : return -1
    else : return 0
    
print(sorted([1,3,5,7,9,8,6,4,2,0], key=cmp_to_key(comparator_ascending)))
print(sorted([1,3,5,7,9,8,6,4,2,0], key=cmp_to_key(comparator_descending)))
```

```python
# binary search
# https://docs.python.org/ko/dev/library/bisect.html
import bisect

lst = [1,2,3,5,7,8,9,10]
bl = bisect.bisect_left(lst, 5)
br = bisect.bisect_right(lst, 5)

bisect.insort_left(lst, 4)
bisect.insort_right(lst, 6)

print(bl, br, b, lst)
```

```python
# heap (min heap)
# https://www.daleseo.com/python-heapq/
import heapq

pq = [1,3,5,7,9,8,6,4,2,0]

heapq.heapify(pq)
heapq.heappush(pq, -1)
p = heapq.heappop(pq)
s = heapq.nsmallest(1, pq)
ss = pq[0]
l = heapq.nlargest(1, pq)

print(lst, p, s, ss, l)
```

```python
# heap with multi-key
import heapq

pq = [(1,1), (2,3), (3,1), (1,2), (2,1), (2,2)]

heapq.heapify(pq)
print(pq)
```

```python
# defaultdict
from collections import defaultdict

dic = defaultdict(lambda : [])

dic[1].append(11)
dic[2].append(22)

print(list(dic.items()))
print(list(dic.values()))
```

```python
# product of number in list
import numpy as np

lst = [1,2,3,4]
np.prod(lst)
```

```python
# (약수, 약수)의 리스트를 가져오기
def get_divisors(n) :
    result = []
    for idx in range(1, int(n**0.5)+1) :
        if n%idx == 0 :
            result.append((idx, n//idx))
    return result

get_divisors(225)
```
